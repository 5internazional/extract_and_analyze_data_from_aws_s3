```python
import pandas as pd
import glob, os
```

```python
#convert in csv
def load_csv_dir(data_dir):
    return pd.concat(map(pd.read_csv, glob.glob(os.path.join(data_dir, '*.csv'))))

data_dir = '/Users/alexandra/Desktop/coding/challenge_dkb/data/deutsche-boerse-xetra-pds/%s/' % date
unprococessed_df = load_csv_dir(data_dir)
```

```python
isin_name = 'AT00000FACC2'
```

```python
#trying with 1 particular ISIN
def extract_info_from_one_isin(df, name, date):
    one_isin = df.loc[(df['ISIN']==name)] 
    start_time = one_isin['Time'].min()
    end_time = one_isin['Time'].max()
    start_price = one_isin.loc[(one_isin['Time']==start_time)]['StartPrice']
    end_price = one_isin.loc[(one_isin['Time']==end_time)]["EndPrice"]
    trade_volum = one_isin['TradedVolume'].sum()
    final_df = pd.DataFrame([{'ISIN':name, 'date':date, 'opening_price':start_price.values[0], 'closing_price':end_price.values[0],  'daily_traded_volume':trade_volum}]) 
    return final_df
```

```python
extract_info_from_one_isin(unprococessed_df, isin_name, date)
```

```python
#transforming and generelazing a method "extract_info_from_one_isin" for multiple ISIN
aggregated = unprococessed_df[['ISIN', 'Time', 'TradedVolume']].groupby(['ISIN']).agg({'Time': ['min', 'max'], 'TradedVolume': 'sum'})
```

```python
aggregated
```

```python
data_merged = unprococessed_df.merge(aggregated, on='ISIN', suffixes=('', '_min', '_max', '_sum'))

```

```python
start_price = data_merged[data_merged.Time==data_merged[('Time', 'min')]][['ISIN', 'StartPrice']].sort_values('ISIN')
```

```python
end_price = data_merged[data_merged.Time==data_merged[('Time', 'max')]][['ISIN', 'EndPrice']].sort_values('ISIN')
```

```python
df = pd.merge(start_price, end_price).merge(aggregated, on='ISIN').rename({"StartPrice": "opening_price", "EndPrice": "closing_price"}, axis="columns")
```

```python
df = df.set_axis([*df.columns[:-1], 'daily_traded_volume'], axis=1, inplace=False)
```

```python
df.rename(columns={ df.columns[4]: "max" }, inplace = True)
df.rename(columns={ df.columns[3]: "min" }, inplace = True)
final_df=df.drop(['min', 'max'], axis=1)
final_df['date'] = date
final_df
```
